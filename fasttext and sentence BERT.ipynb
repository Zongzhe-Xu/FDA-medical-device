```python
import io
import string
import os
os.chdir('C:\\Users\\22575\\Desktop\\xzz\\Doctorate\\')
def LoadFastText(fname):
    input_file = io.open(fname, 'r', encoding='utf-8', newline='\n', errors='ignore')
    no_of_words, vector_size = map(int, input_file.readline().split())
    word_to_vector: Dict[str, List[float]] = dict()
    for i, line in enumerate(input_file):
        tokens = line.rstrip().split(' ')
        word = tokens[0]
        vector = list(map(float, tokens[1:]))
        assert len(vector) == vector_size
        word_to_vector[word] = vector
    return word_to_vector
w2v = LoadFastText('crawl-300d-2M-subword\\crawl-300d-2M-subword.vec')
```

```python
def fast_embed(text,wv):
    text=text.replace("\n", " ")
    text=text.replace("-"," ")
    text=text.lower()
    text="".join(c for c in text if c not in string.punctuation)
    tokens = text.split()
    word_embeddings=np.array([0.0]*300)
    l=0
    for token in tokens:
        token_embedding=wv.get(token, [])
        if len(token_embedding)==0:
            l+=1
            print(token)
            continue
        word_embeddings+=np.array(token_embedding)
    sentence_embedding=word_embeddings/(len(tokens)-l)
    return sentence_embedding.tolist()
```

```python
def cos_similarity(x,y):
    cosine = np.dot(x,y)/(np.linalg.norm(x)*np.linalg.norm(y))
    return cosine
```
